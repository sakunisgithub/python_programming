```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable
```

```python
np.random.seed(15)
```

```python
n_steps = 1000
T = 1.0
dt = T / n_steps
t = np.linspace(0, T, n_steps)
```

```python
configs = [
    (1.2, 0),
    (1.2, 0),
    (1.2, 0),
    (1.8, 0),
    (1.8, 0),
    (1.8, 0)
]
```

```python
plt.figure(figsize=(12, 7))

for alpha, beta in configs:

    increments = levy_stable.rvs(
        alpha,
        beta,
        scale=dt**(1/alpha),
        size=n_steps
    )

    path = np.cumsum(increments)

    plt.plot(
        t,
        path,
        linewidth=2,
        label=fr'$\alpha$={alpha},$\beta$={beta}'
    )


brownian_increments = np.sqrt(dt) * np.random.randn(n_steps)
brownian_path = np.cumsum(brownian_increments)

plt.plot(
    t,
    brownian_path,
    linewidth=2.5,
    label='Brownian Motion'
)

plt.legend(fontsize=12)

plt.tight_layout()

plt.savefig("Brownian Motion and Levy Motions.pdf", bbox_inches="tight")

plt.show()
```
