```python
import random
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
n = int(input("Enter the number of times you want to toss the coin = "))
```

```python
number_of_toss = np.arange(1, n+1)

coin = [0, 1]

toss_results = np.zeros(n)
```

```python
for i in range(n) :
    toss_results[i] = random.choice(coin)
```

```python
frequency_of_heads = np.cumsum(toss_results == 1)
frequency_of_tails = np.cumsum(toss_results == 0)
```

```python
relative_frequency_of_heads = np.zeros(n)
relative_frequency_of_tails = np.zeros(n)
```

```python
for i in range(n) :
    relative_frequency_of_heads[i] = frequency_of_heads[i] / (i+1)
    relative_frequency_of_tails[i] = frequency_of_tails[i] / (i+1)
```

```python
my_dictionary = {
    'No. of Toss' : number_of_toss,
    'Frequency of Head' : frequency_of_heads,
    'Relative Frequency of Head' : relative_frequency_of_heads,
    'Frequency of Tail' : frequency_of_tails,
    'Relative Frequency of Tail' : relative_frequency_of_tails
}

experiment_result = pd.DataFrame(my_dictionary)

print(experiment_result.to_string(index=False))
```

```python
plt.scatter(number_of_toss, relative_frequency_of_heads, s=1, c='black')

plt.xlabel("No. of Tosses")
plt.ylabel("Relative Frequency of Heads")
plt.title("Relative Frequency of Heads vs No. of Tosses")

plt.show()
```

```python
plt.scatter(number_of_toss, relative_frequency_of_tails, s=1, c='black')

plt.xlabel("No. of Tosses")
plt.ylabel("Relative Frequency of Tails")
plt.title("Relative Frequency of Tails vs No. of Tosses")

plt.show()
```
