# Lambda Function in Python

In Python, lambda function (also called an anonymous function) is a small anonymous function that can be defined in a single line of code without a name. It is useful when we need a simple function that we don't want to define explicitely by using the 'def' keyword. The basic syntax for lambda function in python is as follows :

*lambda arguments: expression*

Here, *arguments* refer to the inputs of the function and *expression* is a signle line expression that gets evaluated and returned as the result of the function. There is no need of return statement.

```python
square = lambda x:x**2
```

```python
type(square)
```

```python
print(square(5))
```

```python
sum = lambda x, y:x+y
```

```python
print(sum(2, 3))
```

```python
x = lambda x : "Even" if x % 2 == 0 else "Odd"
```

```python
x(3)
```

```python
x(4)
```

```python
## Filer out even numbers from a list.

a = [1, 2, 3, 4, 5, 6]

even = list(filter(lambda x : x % 2 == 0, a))

print(even)
```

```python
# Get names of students.
students = [
    {
        "name": "Jacob Martin",
        "father name": "Ros Martin",
        "Address": "123 Hill Street",
    },
    {
        "name": "Angela Stevens",
        "father name": "Robert Stevens",
        "Address": "3 Upper Street London",
    },
    {
        "name": "Ricky Smart",
        "father name": "William Smart",
        "Address": "Unknown",
    }
]

names = list(map(lambda x : x["name"], students))
print(names)
```

```python
# filter()

x = [1, 2, 3, 4, 5, 6, 7]

a = list(filter(lambda x : x > 4, x))

print(a)
```

```python
fruits = ['apple', 'mango', 'orange', 'banana']

a = list(filter(lambda x : 'o' in x, fruits))

print(a)
```

```python
# reduce()

a = [1, 2, 3, 4, 5, 6]

from functools import reduce

max_in_a = reduce(lambda x, y : x if x > y else y, a)

print(max_in_a)
```
