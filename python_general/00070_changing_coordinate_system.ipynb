```python
import matplotlib.pyplot as plt
import numpy as np

# Vectors
v = np.array([-1, 3])
Av = np.array([2, 8])

# Set up the plot
fig, ax = plt.subplots()
ax.axhline(0, color='black', linewidth=1)  # x-axis
ax.axvline(0, color='black', linewidth=1)  # y-axis

# Plot the vectors
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')
ax.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1, color='green')

# Axes limits
ax.set_xlim(-2, 5)
ax.set_ylim(-2, 10)

ax.set_title("Vectors in Standard Basis")

plt.show()
```

```python
# Vectors
v = np.array([-1, 3])
Av = np.array([2, 8])

# Set up the plot
fig, ax = plt.subplots()

# Plot the vectors
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')
ax.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1, color='green')

x_vals = np.linspace(-10, 10, 400)
y1 = -x_vals  # x + y = 0
y2 = x_vals / 2  # x - 2y = 0 → y = x/2

ax.plot(x_vals, y1, label='x + y = 0')      # eigenvector direction (1, -1)
ax.plot(x_vals, y2, 'purple', label='x - 2y = 0')  # eigenvector direction (2, 1)

# Axes limits
ax.set_xlim(-2, 5)
ax.set_ylim(-2, 10)

ax.set_title("Vectors in Eigen-Basis")

plt.show()
```
